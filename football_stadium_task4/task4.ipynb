```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
runs = ["no priority", "priority\nqueues", "priority &\n0.5h delay", "priority &\n1h delay"]
#runs = ["(a)", "(b)", "(c)", "(d)"]

entering_after_kickoff = [78.0, 119.1, 152.0, 805.7]
check_ins_per_hour = [3010.5, 2988.8, 3011.9, 2703.5]

mean_waiting_time = [[31.3, 23.0], [59.5, 22.1], [59.9, 23.9], [92.6, 52.5]]
mean_waiting_time = pd.DataFrame(mean_waiting_time, index=runs, columns=["standard", "season ticket"]).reset_index(names="variant")
mean_waiting_time = pd.melt(mean_waiting_time, id_vars="variant", var_name="ticket type", value_name="waiting time [s]")

```

```python
fig_width = 6
fig_heigt = 3
sns.set(style="darkgrid")
fig, ax = plt.subplots(1, 3, figsize=(3 * fig_width, 1 * fig_heigt))
_ = sns.barplot(x=runs, y=entering_after_kickoff, ax=ax[0]).set_title("fans entering stadium after kickoff")
_ = sns.barplot(x=runs, y=check_ins_per_hour, ax=ax[1]).set_title("mean check-ins per hour")
_ = (sns.barplot(x='variant', y='waiting time [s]', hue='ticket type', data=mean_waiting_time, ax=ax[2])
     .set(xlabel='', ylabel='', title='mean waiting time in seconds'))
```
